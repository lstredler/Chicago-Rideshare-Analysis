```python
import pandas as pd
import requests
from config import api_key

march_trips_file = "march2020.csv"
df = pd.read_csv(march_trips_file)
api_key
```

```python
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head()
```

```python
df_locations = df[['pickup_community_area','dropoff_community_area','pickup_centroid_latitude','pickup_centroid_longitude','dropoff_centroid_latitude','dropoff_centroid_longitude']]
df_locations_clean = df_locations.dropna()
df_locations_clean = df_locations_clean.astype({'pickup_community_area':'int32','dropoff_community_area':'int32'})
df_locations_clean.dtypes
```

```python
areas = requests.get('https://data.cityofchicago.org/resource/igwz-8jzy.json').json()
areas_df = pd.DataFrame(areas)[['community','area_num_1']]
areas_dict = dict(zip(areas_df.area_num_1, areas_df.community))
pd.DataFrame(areas).iloc[0][0]
```

```python
df_locations_clean['pickup_area'] = df_locations_clean['pickup_community_area'].map(lambda x: areas_dict[str(x)])
df_locations_clean['dropoff_area'] = df_locations_clean['dropoff_community_area'].map(lambda x: areas_dict[str(x)])
df_locations_clean
```

```python
# top 30 most popular pickup areas
top_pickup_areas = df_locations_clean.pickup_area.value_counts()
top_pickup_areas.head(15).plot.bar()
top_pickup_areas
```

```python
# top 30 most popular dropoff areas
top_dropoff_areas = df_locations_clean.dropoff_area.value_counts().head(15)
top_dropoff_areas.plot.bar()
```

```python
popular_trips_df = df_locations_clean.groupby(['pickup_area','dropoff_area']).size().reset_index(name='count').sort_values(by='count', ascending=False).head(15)
popular_trips_df['trip'] = popular_trips_df.pickup_area + ' to ' + popular_trips_df.dropoff_area
popular_trips_df = popular_trips_df[['trip','count']].set_index('trip')
popular_trips_df.plot.bar();
```

```python
pickup_coord_df = df_locations_clean.groupby(['pickup_centroid_latitude','pickup_centroid_longitude','pickup_area']).size().reset_index(name='count').sort_values(by='count', ascending=False)
pickup_coord_df
# popular_trips_df['trip'] = popular_trips_df.pickup_area + ' to ' + popular_trips_df.dropoff_area
# popular_trips_df = popular_trips_df[['trip','count']].set_index('trip')
# popular_trips_df.plot.bar();
```

```python
import gmaps
gmaps.configure(api_key=api_key)
locations = pickup_coord_df[['pickup_centroid_latitude','pickup_centroid_longitude']]
weights = pickup_coord_df['count']
fig = gmaps.figure(center=(41.8781,-87.6298),zoom_level=12)
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights))
fig
```

```python
top_pickup_coord_df = pickup_coord_df.head(20)
info_box_template = """
<dl>
<dt>Area</dt><dd>{pickup_area}</dd>
<dt>pickups</dt><dd>{count}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
pickup_info = [info_box_template.format(**row) for index, row in top_pickup_coord_df.iterrows()]
locations = top_pickup_coord_df[["pickup_centroid_latitude", "pickup_centroid_longitude"]].head(20)
```

```python
# Add marker layer ontop of heat map
marker_layer = gmaps.marker_layer(
    locations,
    info_box_content = pickup_info
)
fig.add_layer(marker_layer)

# Display figure
fig
```

```python
import folium
m = folium.Map(location=[41.8781,-87.6298])
bins = list(top_pickup_areas.quantile([0, 0.25, 0.5, 0.75, 1]))
folium.Choropleth(
    geo_data='chicago-community-areas.geojson',
    name="choropleth",
    data=top_pickup_areas,
    key_on='feature.properties.community',
    bins=bins
).add_to(m)
m
```
