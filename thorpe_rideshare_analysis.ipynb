```python
original_df['day'] = pd.DatetimeIndex(original_df['Trip Start Timestamp']).day
original_df
```

```python
day_counts = original_df["day"].value_counts().reset_index().sort_values("index")
day_counts
```

```python
df_group = original_df.groupby(original_df["day"])

df_group
```

```python
thorpe_bar = day_counts.plot.bar("index","day", xlabel = 'Day of the Month', ylabel = 'Number of Rides Per Day')
thorpe_bar
```

```python
thorpe_line = day_counts.plot.line("index","day", xlabel = 'Day of the Month', ylabel = 'Number of Rides Per Day')
thorpe_line
```
